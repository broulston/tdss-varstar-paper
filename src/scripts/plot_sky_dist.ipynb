```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.table import Table
from astropy import units as u
from astropy.coordinates import SkyCoord

import paths
```

```python
# data = Table.read('/Users/benjaminroulston/Dropbox/Research/Projects/Variable_Stars/HARD_COPY_ORGINAL_DATA/PROGRAM_SAMPLE/2021-07-27/FINAL_FILES/TDSS_SES+PREV_DR16DR12griLT20_GaiaEDR3_Drake2014PerVar_CSSID_ZTFIDs_LCpointer_PyHammer_EqW_LCProps.fits')

data = Table.read(paths.data / 'TDSS_VarStar_FINAL_Var_ALL_PROP_STATS_dealiased.fits')
```

```python
c = SkyCoord(ra=data['ra_GaiaEDR3'], dec=data['dec_GaiaEDR3'], frame='icrs')
ra_rad = c.ra.wrap_at(180 * u.deg).radian
dec_rad = c.dec.radian
```

```python
plt.figure(figsize=(10,5), dpi=600)
plt.subplot(111, projection="aitoff")
ax = plt.gca()
# plt.title("Aitoff projection of our random data")
ax.grid(True)
ax.plot(ra_rad, dec_rad, c='k', marker='.', markersize=1, ls='None', alpha=0.3)
# plt.subplots_adjust(top=0.95,bottom=0.0)
ax.set_xticklabels(['14h','16h','18h','20h','22h','0h','2h','4h','6h','8h','10h'])
plt.tight_layout()
plt.savefig(paths.figures / "TDSS_VarStar_Sample_SkyDist.pdf", dpi=600)
```
